```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data = pd.read_excel('./data.xls')
data.head()
```

```python
data.info()
```

```python
df = pd.DataFrame(data)
df.info()
```

```python
# Ubah data transaksional menjadi format yang sesuai untuk Apriori
df_encoded = pd.get_dummies(df.apply(lambda x: pd.Series(x.dropna()), axis=1)).astype(bool)

# Terapkan algoritma Apriori untuk menemukan itemset yang sering muncul
frequent_itemsets = apriori(df_encoded, min_support=0.10  , use_colnames=True)


print(frequent_itemsets)
```

```python
rules = association_rules(frequent_itemsets,metric="lift",min_threshold=1)

list_data = list(frequent_itemsets)
```

```python
print(list_data)
```

```python
rules.describe()
```

```python
rules.to_excel('data.xlsx',index=False)
```
